```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

load the data

```python
df= pd.read_csv("C:\\Ola_Uber_booking_project\\Bookings.csv")
df.head()
```

```python
df.describe()
```

checking for null values

```python
df.isnull().sum()
```

Drop garbage or unwanted columns

```python
df.drop(columns=['Unnamed: 20','Vehicle Images'],inplace=True)
```

convert the date and time data type

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S')
```

```python
df.info()
```

Creating new columns

```python
df['Hour'] = pd.to_datetime(df['Time'], format= '%H:%M:%S').dt.hour

df['Is_Canceled'] = df['Booking_Status'].apply(
    lambda x: 1 if 'Cancel' in x else 0
)
```

Handle missing values

```python
df['Driver_Ratings'].fillna(df['Driver_Ratings'].median(),inplace=True)
df['Customer_Rating'].fillna(df['Customer_Rating'].median(),inplace=True)

df['Payment_Method'].fillna('Unknown',inplace=True)
```

```python
df.isnull().sum()
```

Feature Engineering

```python
df['Ride_Type'] = pd.cut(
    df['Ride_Distance'],
    bins=[0,5,15,30,100],
    labels=['Short','Medium','Long','Very Long']
)
```

```python
total_bookings = df['Booking_ID'].nunique()
total_revenue = df['Booking_Value'].sum()
cancel_rate = df['Is_Canceled'].mean() * 100

print('total booking= ',total_bookings,'total revenue= ',total_revenue,'cancel rate= ',cancel_rate)
```

```python
df.groupby('Vehicle_Type')['Is_Canceled'].mean().sort_values(ascending=False)
```
